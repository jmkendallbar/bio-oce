````python
# Canonical names — species where NCBI has a binomial entry; genus name where the
# table only identifies to genus level (NCBI will place them at genus rank in the tree).
SPECIES = [
    # Phytoplankton
    'Pseudo-nitzschia',               # genus only in identification table
    #'Eucyrtidium',  # genus of Radiolarian
    #'',
    'Chaetoceros',                    # genus only
    'Alexandrium catenella',
    'Dinophysis',                     # genus only
    'Gambierdiscus',                  # genus only
    'Karenia brevis',
    'Margalefidinium',                # genus only; formerly Cochlodinium
    'Dolichospermum',                 # genus only
    'Tripos',                         # genus only; formerly Ceratium
    'Emiliania huxleyi',              # fallback: 'Gephyrocapsa huxleyi'
    # 'Coccolithus pelagicus',
    # Food web / outgroup taxa
    'Oikopleura dioica',
    'Calanus pacificus',
    'Sardinops sagax',
    'Engraulis mordax',
    'Metacarcinus magister',
    'Zalophus californianus',
    'Mirounga angustirostris',
    'Homo sapiens',
    'Pycnopodia helianthoides',
]

# Name corrections for synonyms/basionyms — map original → NCBI-accepted name
NAME_CORRECTIONS: dict[str, str] = {
    # Uncomment if needed:
    # 'Emiliania huxleyi': 'Gephyrocapsa huxleyi',
}

print(f'{len(SPECIES)} names defined')
for s in SPECIES:
    print(' -', s)
````

## 2. Build induced taxonomy tree with ete3

````python
ncbi = NCBITaxa(dbfile=str(LOCAL_DB))

# Apply synonym corrections
species_remapped = [NAME_CORRECTIONS.get(s, s) for s in SPECIES]
corrected = [(orig, new) for orig, new in zip(SPECIES, species_remapped) if orig != new]
for orig, new in corrected:
    print(f'  corrected: {orig!r} → {new!r}')

# Map names → taxids (mix of species and genus names — ete3 handles both)
name_to_taxid = ncbi.get_name_translator(species_remapped)

missing = [s for s in species_remapped if s not in name_to_taxid]
if missing:
    print('⚠ Could not resolve (skipping):')
    for m in missing:
        print(' -', m)
    print('Add a fallback synonym to NAME_CORRECTIONS above and re-run.')

taxids = [name_to_taxid[s][0] for s in species_remapped if s in name_to_taxid]

# Induced topology
ete_tree = ncbi.get_topology(taxids)

# Relabel leaves: taxid → scientific name
taxid_to_name = ncbi.get_taxid_translator([int(leaf.name) for leaf in ete_tree.iter_leaves()])
for leaf in ete_tree.iter_leaves():
    leaf.name = taxid_to_name[int(leaf.name)]

print(ete_tree.get_ascii(show_internal=True))
````

## 3. Export to Newick (with NHX taxonomic annotations)

Each node gets `[&&NHX:taxid=...:phylum=...:class=...:order=...]` comments embedded in the Newick string. Leaf nodes get their own lineage; internal nodes use the NCBI taxid stored by ete3.

````python
import re as _re

RANKS_WANTED = ('phylum', 'class', 'order')

def _lineage_ranks(taxid: int) -> dict[str, str]:
    """Return {rank: name} for the ranks we care about, from a taxid's lineage."""
    lineage = ncbi.get_lineage(taxid)
    ranks   = ncbi.get_rank(lineage)
    names   = ncbi.get_taxid_translator(lineage)
    result  = {}
    for tid in lineage:
        r = ranks.get(tid, '')
        if r in RANKS_WANTED:
            result[r] = names[tid].replace(' ', '_').replace(',', '')
    return result

def _nhx(node) -> str:
    """Build a &&NHX comment for an ete3 node."""
    try:
        tid = int(node.name)
    except (ValueError, AttributeError):
        matches = name_to_taxid.get(node.name, [])
        if not matches:
            return ''
        tid = matches[0]
    parts = [f'taxid={tid}']
    for rank, val in _lineage_ranks(tid).items():
        parts.append(f'{rank}={val}')
    return '[&&NHX:' + ':'.join(parts) + ']'

plain_newick = ete_tree.write(format=1)

node_nhx: dict[str, str] = {}
for node in ete_tree.traverse():
    nhx = _nhx(node)
    if nhx:
        node_nhx[node.name] = nhx

def _inject_nhx(newick: str, nhx_map: dict[str, str]) -> str:
    for name in sorted(nhx_map, key=len, reverse=True):
        nhx = nhx_map[name]
        escaped = _re.escape(name)
        newick = _re.sub(rf'({escaped})(?=[,):;])', rf'\1{nhx}', newick)
    return newick

nhx_newick = _inject_nhx(plain_newick, node_nhx)
newick_str = nhx_newick   # keep alias so downstream cells work unchanged

(OUTPUT_DIR / 'phylogeny').mkdir(parents=True, exist_ok=True)
OUT_NEWICK.write_text(nhx_newick, encoding='utf-8')
print('Wrote annotated Newick to', OUT_NEWICK)
print()
print(nhx_newick[:800], '...')
````

## 6. Export iTOL family annotation (DATASET_COLORSTRIP)

````python
import colorsys

OUT_ITOL_FAMILY = OUT_ITOL / 'itol_family_colorstrip.txt'

# --- 1. resolve family for each leaf species ---
leaf_names = [leaf.name for leaf in ete_tree.iter_leaves()]
leaf_taxids = {name: name_to_taxid[name][0] for name in leaf_names if name in name_to_taxid}

def get_rank_name(taxid: int, rank: str) -> str | None:
    """Walk the lineage and return the name at *rank*, or None."""
    lineage = ncbi.get_lineage(taxid)
    ranks   = ncbi.get_rank(lineage)
    names   = ncbi.get_taxid_translator(lineage)
    for tid in lineage:
        if ranks.get(tid) == rank:
            return names[tid]
    return None

species_to_family: dict[str, str] = {}
for name, tid in leaf_taxids.items():
    family = get_rank_name(tid, 'family') or 'Unknown'
    species_to_family[name] = family

families = sorted(set(species_to_family.values()))
print(f'Families ({len(families)}):', families)

# --- 2. assign a distinct colour per family ---
def make_palette(n: int) -> list[str]:
    colours = []
    for i in range(n):
        h = i / n
        r, g, b = colorsys.hsv_to_rgb(h, 0.65, 0.85)
        colours.append(f'#{int(r*255):02X}{int(g*255):02X}{int(b*255):02X}')
    return colours

palette = make_palette(len(families))
family_colour = dict(zip(families, palette))

# --- 3. write iTOL DATASET_COLORSTRIP file ---
lines = [
    'DATASET_COLORSTRIP',
    'SEPARATOR TAB',
    'DATASET_LABEL\tFamily',
    'COLOR\t#888888',
    'STRIP_WIDTH\t40',
    'SHOW_INTERNAL\t0',
    'LEGEND_TITLE\tFamily',
    'LEGEND_SHAPES\t' + '\t'.join('1' for _ in families),
    'LEGEND_COLORS\t' + '\t'.join(family_colour[f] for f in families),
    'LEGEND_LABELS\t' + '\t'.join(families),
    'DATA',
]
for name in leaf_names:
    family = species_to_family.get(name, 'Unknown')
    colour = family_colour.get(family, '#888888')
    lines.append(f'{name}\t{colour}\t{family}')

OUT_ITOL_FAMILY.write_text('\n'.join(lines) + '\n', encoding='utf-8')
print('Wrote iTOL annotation to', OUT_ITOL_FAMILY)
````

````python
# ── Reorder family palette by phylogenetic distance ───────────────────────
import numpy as np
from collections import defaultdict
from scipy.cluster.hierarchy import linkage, leaves_list
from scipy.spatial.distance import squareform

family_to_leaves: dict[str, list[str]] = defaultdict(list)
for name, fam in species_to_family.items():
    family_to_leaves[fam].append(name)

fam_list = families
n_fam = len(fam_list)
dist_matrix = np.zeros((n_fam, n_fam))

for i, fi in enumerate(fam_list):
    for j, fj in enumerate(fam_list):
        if i >= j:
            continue
        dists = [
            ete_tree.get_distance(a, b)
            for a in family_to_leaves[fi]
            for b in family_to_leaves[fj]
        ]
        mean_d = float(np.mean(dists)) if dists else 0.0
        dist_matrix[i, j] = mean_d
        dist_matrix[j, i] = mean_d

condensed = squareform(dist_matrix, checks=False)
if n_fam > 1:
    Z = linkage(condensed, method='average')
    order = leaves_list(Z)
else:
    order = [0]

families_ordered = [fam_list[i] for i in order]
print('Families ordered by phylogenetic proximity:')
for f in families_ordered:
    print(' ', f)

palette_ordered = make_palette(n_fam)
family_colour = {fam: palette_ordered[rank] for rank, fam in enumerate(families_ordered)}

# Overwrite DATASET_COLORSTRIP with phylogenetically-ordered colours
lines = [
    'DATASET_COLORSTRIP',
    'SEPARATOR TAB',
    'DATASET_LABEL\tFamily (phylo order)',
    'COLOR\t#888888',
    'STRIP_WIDTH\t40',
    'SHOW_INTERNAL\t0',
    'LEGEND_TITLE\tFamily',
    'LEGEND_SHAPES\t' + '\t'.join('1' for _ in families_ordered),
    'LEGEND_COLORS\t' + '\t'.join(family_colour[f] for f in families_ordered),
    'LEGEND_LABELS\t' + '\t'.join(families_ordered),
    'DATA',
]
for name in leaf_names:
    family = species_to_family.get(name, 'Unknown')
    colour = family_colour.get(family, '#888888')
    lines.append(f'{name}\t{colour}\t{family}')

OUT_ITOL_FAMILY.write_text('\n'.join(lines) + '\n', encoding='utf-8')
print('Overwrote iTOL annotation with phylo-ordered colours:', OUT_ITOL_FAMILY)
````

## 7. Export iTOL Order colorrange (DATASET_RANGE)

````python
import re

# 1. Resolve order for each leaf
species_to_order: dict[str, str] = {}
for name, tid in leaf_taxids.items():
    order = get_rank_name(tid, 'order') or 'Unknown'
    species_to_order[name] = order

orders = sorted(set(species_to_order.values()))
print(f'Orders ({len(orders)}):', orders)

# 2. Build order x order mean-distance matrix
order_to_leaves: dict[str, list[str]] = defaultdict(list)
for name, ord_ in species_to_order.items():
    order_to_leaves[ord_].append(name)

n_ord = len(orders)
ord_dist = np.zeros((n_ord, n_ord))
for i, oi in enumerate(orders):
    for j, oj in enumerate(orders):
        if i >= j:
            continue
        dists = [
            ete_tree.get_distance(a, b)
            for a in order_to_leaves[oi]
            for b in order_to_leaves[oj]
        ]
        mean_d = float(np.mean(dists)) if dists else 0.0
        ord_dist[i, j] = mean_d
        ord_dist[j, i] = mean_d

# 3. Hierarchical clustering → dendrogram order
condensed_ord = squareform(ord_dist, checks=False)
if n_ord > 1:
    Z_ord = linkage(condensed_ord, method='average')
    ord_order = leaves_list(Z_ord)
else:
    ord_order = [0]

orders_ordered = [orders[i] for i in ord_order]
print('Orders by phylogenetic proximity:', orders_ordered)

palette_orders = make_palette(n_ord)
order_colour = {ord_: palette_orders[rank] for rank, ord_ in enumerate(orders_ordered)}

# 4. Build order → list of (node_id, label) entries
internal_ids_in_newick = set(re.findall(r'\)(\d+):', newick_str))
leaf_names_list = [leaf.name for leaf in ete_tree.iter_leaves()]

def get_order_entries(ord_: str) -> list[tuple[str, str]]:
    leaves_in_order = set(order_to_leaves[ord_])
    if len(leaves_in_order) == 1:
        leaf = next(iter(leaves_in_order))
        return [(leaf, ord_)]
    lca = ete_tree.get_common_ancestor(list(leaves_in_order))
    lca_leaves = {l.name for l in lca.iter_leaves()}
    if lca_leaves == leaves_in_order and lca.name in internal_ids_in_newick:
        return [(lca.name, ord_)]
    else:
        entries = []
        for child in lca.children:
            child_order_leaves = {l.name for l in child.iter_leaves()} & leaves_in_order
            if not child_order_leaves:
                continue
            if len(child_order_leaves) == 1:
                entries.append((next(iter(child_order_leaves)), ''))
            elif child.name in internal_ids_in_newick:
                entries.append((child.name, ''))
            else:
                sub_ordered = [n for n in leaf_names_list if n in child_order_leaves]
                entries.append((sub_ordered[0], ''))
        if entries:
            entries[-1] = (entries[-1][0], ord_)
        return entries

order_entries: dict[str, list[tuple[str, str]]] = {}
for ord_ in orders_ordered:
    entries = get_order_entries(ord_)
    order_entries[ord_] = entries
    for nid, label in entries:
        print(f'  {ord_:<25} node={nid:<40} label={label!r}')

# 5. Write DATASET_RANGE
lines = [
    'DATASET_RANGE',
    'SEPARATOR COMMA',
    'DATASET_LABEL,Order',
    'COLOR,#888888',
    '',
    'RANGE_TYPE,box',
    'RANGE_COVER,clade',
    'SHOW_LABELS,1',
    'LABEL_POSITION,bottom-right',
    'COVER_LABELS,0',
    'COVER_DATASETS,0',
    '',
    'LEGEND_TITLE,Order',
    'LEGEND_SHAPES,' + ','.join('1' for _ in orders_ordered),
    'LEGEND_COLORS,' + ','.join(order_colour[o] for o in orders_ordered),
    'LEGEND_LABELS,' + ','.join(orders_ordered),
    '',
    '#START_NODE_ID,END_NODE_ID,FILL_COLOR,GRADIENT_COLOR,LINE_COLOR,LINE_STYLE,LINE_WIDTH,LABEL_TEXT,LABEL_COLOR,LABEL_SIZE_FACTOR,LABEL_STYLE',
    'DATA',
]

for ord_ in orders_ordered:
    col = order_colour.get(ord_, '#888888')
    for nid, label in order_entries[ord_]:
        lines.append(f'{nid},{nid},{col},,{col},solid,1,{label},#ffffff,1,normal')

OUT_ITOL_ORDER = OUT_ITOL / 'itol_order_colorrange.txt'
OUT_ITOL_ORDER.write_text('\n'.join(lines) + '\n', encoding='utf-8')
print('Wrote iTOL order colorrange to', OUT_ITOL_ORDER)
````

## 8. PhyloPic silhouettes + common names → iTOL DATASET_IMAGE & DATASET_TEXT

iTOL resolves internal nodes by **leaf-pair notation** (`leaf1|leaf2`). Common names come from a hardcoded leaf map plus a clade-label dict keyed by NCBI taxid — written as a separate `DATASET_TEXT` file so labels appear next to each node/leaf.

````python
import json as _json
import time
import urllib.request
import urllib.error

PHYLOPIC_BASE = 'https://api.phylopic.org'
CACHE_PATH = OUTPUT_DIR / 'phylogeny' / 'phylopic_cache.json'

if CACHE_PATH.exists():
    with open(CACHE_PATH) as _f:
        _cache: dict[str, str | None] = _json.load(_f)
else:
    _cache = {}

def _get_json(url: str) -> dict:
    req = urllib.request.Request(url, headers={'Accept': 'application/json'})
    with urllib.request.urlopen(req, timeout=10) as r:
        return _json.loads(r.read())

def _fetch_image_for_taxid(taxid: int) -> str | None:
    key = str(taxid)
    if key in _cache:
        return _cache[key]
    try:
        resolve_url = (
            f'{PHYLOPIC_BASE}/resolve/ncbi.nlm.nih.gov/taxid/{taxid}'
            '?embed%5BprimaryImage%5D=true'
        )
        node_data = _get_json(resolve_url)
        img_href = node_data.get('_links', {}).get('primaryImage', {}).get('href', '')
        if not img_href:
            _cache[key] = None
            return None
        img_uuid = img_href.rstrip('/').split('/')[-1].split('?')[0]
        raster_url = f'{PHYLOPIC_BASE}/images/{img_uuid}?embed%5BrasterFiles%5D=true'
        raster_files = _get_json(raster_url).get('_links', {}).get('rasterFiles', [])
        png_url = raster_files[-1].get('href', '') if raster_files else ''
        _cache[key] = png_url or None
        time.sleep(0.12)
        return _cache[key]
    except urllib.error.HTTPError:
        _cache[key] = None
        return None

SKIP_TIDS = {1, 131567}

def _phylopic_url_for_taxid(taxid: int) -> tuple[str | None, str]:
    url = _fetch_image_for_taxid(taxid)
    if url:
        return url, 'exact'
    try:
        lineage = ncbi.get_lineage(taxid)
        ranks   = ncbi.get_rank(lineage)
        names   = ncbi.get_taxid_translator(lineage)
        for ancestor_tid in reversed(lineage[:-1]):
            if ancestor_tid in SKIP_TIDS:
                continue
            url = _fetch_image_for_taxid(ancestor_tid)
            if url:
                ancestor_name = names.get(ancestor_tid, str(ancestor_tid))
                ancestor_rank = ranks.get(ancestor_tid, '?')
                return url, f'ancestor:{ancestor_rank}:{ancestor_name}'
    except Exception as e:
        print(f'    lineage error for {taxid}: {e}')
    return None, 'not found'

# ── Collect (iTOL_id, taxid) for every node ──────────────────────────────────
# Leaves   → leaf name (e.g. "Karenia brevis")
# Internals → "leaf1|leaf2" notation: pick first + last leaf of the clade.
#             iTOL resolves this as the LCA regardless of NHX or node labels.

leaf_name_set = {node.name for node in ete_tree.iter_leaves()}

all_nodes: list[tuple[str, int]] = []

for node in ete_tree.traverse():
    if node.is_leaf():
        tid_list = name_to_taxid.get(node.name)
        if tid_list:
            all_nodes.append((node.name, tid_list[0]))
    else:
        try:
            tid = int(node.name)
        except (ValueError, TypeError):
            continue
        leaves = [l.name for l in node.iter_leaves()]
        if len(leaves) >= 2:
            itol_id = f'{leaves[0]}|{leaves[-1]}'
        else:
            itol_id = leaves[0]   # single-child internal — same as leaf
        all_nodes.append((itol_id, tid))

print(f'Fetching PhyloPic for {len(all_nodes)} nodes '
      f'({sum(n.is_leaf() for n in ete_tree.traverse())} leaves + '
      f'{sum(not n.is_leaf() for n in ete_tree.traverse())} internals) ...')

image_map: dict[str, str] = {}

for itol_id, tid in all_nodes:
    url, source = _phylopic_url_for_taxid(tid)
    if url:
        image_map[itol_id] = url
        tag = '' if source == 'exact' else f'  [{source}]'
        print(f'  OK  {itol_id}{tag}')
    else:
        print(f'  --  {itol_id} (taxid={tid}): not found')

CACHE_PATH.parent.mkdir(parents=True, exist_ok=True)
with open(CACHE_PATH, 'w') as _f:
    _json.dump(_cache, _f, indent=2)
print(f'\nCached {len(_cache)} entries → {CACHE_PATH}')
print(f'Found images for {len(image_map)}/{len(all_nodes)} nodes')

# ── Write iTOL DATASET_IMAGE ──────────────────────────────────────────────────
# position=-1 → outside leaf label; position=1 → at internal node circle
lines = [
    'DATASET_IMAGE',
    'SEPARATOR COMMA',
    'DATASET_LABEL,Silhouettes (PhyloPic)',
    'COLOR,#000000',
    'MARGIN,5',
    'SHOW_INTERNAL,1',
    'SIZE_FACTOR,1',
    'DATA',
]

for itol_id, url in image_map.items():
    is_leaf = '|' not in itol_id and itol_id in leaf_name_set
    pos = '-1' if is_leaf else '1'
    size = '1' if is_leaf else '0.6'
    lines.append(f'{itol_id},{pos},{size},0,0,0,{url}')

OUT_ITOL_IMG = OUT_ITOL / 'itol_phylopic_images.txt'
OUT_ITOL_IMG.write_text('\n'.join(lines) + '\n', encoding='utf-8')
print(f'Wrote {OUT_ITOL_IMG}  ({len(image_map)} entries)')
````
